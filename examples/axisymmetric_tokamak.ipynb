```julia
nr = 100
nz = 120
nl = 10
;
```

```julia
rgrid = LinRange( 0.25,  2.75, nr)
zgrid = LinRange(-2.0,  +2.0,  nz)
;
```

```julia
field = zeros(nr, nz)
potAR = zeros(nr, nz)
potAZ = zeros(nr, nz)
potAP = zeros(nr, nz)
;
```

```julia
for i in 1:nr
    for j in 1:nz
        field[i,j] = B(rgrid[i], zgrid[j], 0.0)
        potAR[i,j] = A₁(rgrid[i], zgrid[j], 0.0)
        potAZ[i,j] = A₂(rgrid[i], zgrid[j], 0.0)
        potAP[i,j] = A₃(rgrid[i], zgrid[j], 0.0)
    end
end
```

```julia
contour(rgrid, zgrid, field', levels=nl, linewidth=0, fillrange=true)
```

```julia
contour(rgrid, zgrid, potAR', levels=nl)
```

```julia
contour(rgrid, zgrid, potAZ', levels=nl)
```

```julia
contour(rgrid, zgrid, potAP', levels=100)
```
